```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
d1 = pd.read_csv("DDQN_T1_Acrobot_0308.csv")
d2 = pd.read_csv("DDQN_T2_Acrobot_0430.csv")
rewards_mean_type1 = d1["Mean"]
reward_std_type1 = d1["Std"]
plt.figure(figsize=[8,5])
plt.plot(np.arange(1,len(rewards_mean_type1)+1),rewards_mean_type1)
plt.fill_between(np.arange(1,len(rewards_mean_type1)+1),rewards_mean_type1 - reward_std_type1,rewards_mean_type1 + reward_std_type1, alpha=0.2)
rewards_mean_type1 = d2["Mean"]
reward_std_type1 = d2["Std"]
plt.plot(np.arange(1,len(rewards_mean_type1)+1),rewards_mean_type1)
plt.fill_between(np.arange(1,len(rewards_mean_type1)+1),rewards_mean_type1 - reward_std_type1,rewards_mean_type1 + reward_std_type1, alpha=0.2)
plt.xlabel("Episodes",fontsize=11)
plt.ylabel(" Rewards",fontsize=11)
plt.title("Rewards Plot",fontsize=13)
plt.legend(["type 1 mean","type 1 std","type 2 mean","type 2 std"])
plt.savefig("DDQN_ACROBOT.png")
plt.show()
```
